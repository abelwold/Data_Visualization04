```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def problem1(file):
    # Read data
    data = pd.read_csv(file)
    # Method for ytick-labels 
    def value(x, pos):
        if x >= 1e6:
            s = '{:1.1f}M'.format(x*1e-6)
        else:
            s = '{:1.0f}K'.format(x*1e-3)
        return s
    # Plot stacked area
    ax = data.plot(kind='area', figsize=(10, 7),colormap = "tab10")
    ax = plt.gca()
    # Change location of legend 
    ax.legend(loc='upper left',frameon=False )
    # Remove first tick-label for the yaxis
    ax.yaxis.get_major_ticks()[0].label1.set_visible(False)
    # Adjust the lable size for the ticks
    ax.tick_params(axis='both', which='major', labelsize=14)
    # Using the defined method assign the tick-labels
    ax.yaxis.set_major_formatter(value)
    ax.set_xlabel("Year", fontsize=20)
    ax.set_ylabel("Unemployment", fontsize=20)
    plt.show()
```

```python
problem1("monthly_unemployment_by_ages.csv")
```
